```python
import matplotlib.pyplot as plt
from numpy import random 
import numpy as np
import sklearn.decomposition as sk
```

```python
n = 1000
w=30
h=4
colors_bw = [tuple([random.rand()]*3) for i in range(n)]
colors = [(random.rand(),random.rand(),random.rand()) for i in range(n)]

# print(colors)
x = np.linspace(0,1,40)
fig, (ax1,ax2) = plt.subplots(2,sharex=True,figsize=(w,h))
ax1.axis('off')
ax1.set_title("RGB Colors (Unsorted)")
ax2.set_title("Grayscale intensities (Unsorted)")
ax2.axis('off')
for i in range(n):
    ax1.bar(i,10,color=colors[i])
    ax2.bar(i,10,color=colors_bw[i])
```

```python
trans_rgb = sk.PCA(n_components=1)
trans_rgb = trans_rgb.fit_transform(colors)
trans_bw = sk.PCA(n_components=1)
trans_bw = trans_bw.fit_transform(colors_bw)
seq_rgb = []
seq_bw = []

for i in range(n):
    seq_rgb.append((i,float(trans_rgb[i])))
    seq_rgb.append((i,float(trans_bw[i])))


fig, (ax1,ax2) = plt.subplots(2,sharex=True,figsize=(w,h))
ax1.axis('off')
ax1.set_title("RGB Colors (Sorted)")
ax2.set_title("Grayscale intensities (Sorted)")
ax2.axis('off')

seq_rgb.sort(key=lambda color: color[1])
seq_bw.sort(key=lambda color: color[1])

plt.axis('off')
for i in range(n):
    index = seq_rgb[i][0]
    index2 = seq_bw[i][0]
    print(index2)
    ax1.bar(i,10,color=colors[index])
    ax2.bar(i,10,color=colors_bw[index2])
```
