```python
import matplotlib.pyplot as plt
from numpy import random 
import numpy as np
import sklearn.decomposition as sk
```

```python
n = 100
w=20
h=2
plt.figure(figsize=(w,h))
# colors = [tuple([random.rand()]*3) for i in range(n)]
colors = [(random.rand(),random.rand(),random.rand()) for i in range(n)]

# print(colors)
x = np.linspace(0,1,40)
plt.axis('off')
for i in range(n):
    plt.bar(i,10,color=colors[i])
```

```python
trans = sk.PCA(n_components=1)
trans = trans.fit_transform(colors)
seq = []
for i in range(n):
    seq.append((i,float(trans[i])))
plt.figure(figsize=(w,h))

seq.sort(key=lambda color: color[1])
plt.axis('off')
for i in range(n):
    index = seq[i][0]
    plt.bar(i,10,color=colors[index])
```
